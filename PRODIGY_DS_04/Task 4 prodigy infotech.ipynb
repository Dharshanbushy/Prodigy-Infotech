## Exploratory Data Analysis of Traffic Accident Data

#### Introduction:

The dataset provided contains detailed information about traffic accidents, including factors such as road conditions, weather, time of day, and severity of accidents. This analysis aims to uncover patterns and insights from the data to better understand the dynamics of traffic accidents and potentially identify areas for intervention or improvement in road safety measures.

The dataset includes various attributes such as Accident_Index, Accident Date, Day_of_Week, Junction_Control, Accident_Severity, Latitude, Longitude, Number_of_Casualties, Number_of_Vehicles, Road_Surface_Conditions, Speed_limit, Time, Urban_or_Rural_Area, Weather_Conditions, Vehicle_Type, among others.

Through exploratory data analysis (EDA), we will visualize the distribution of accidents by different factors such as day of the week, weather conditions, and time of day. Additionally, we will identify accident hotspots based on geographical coordinates.

By conducting this analysis, we aim to gain insights into the underlying patterns and contributing factors of traffic accidents, which can be valuable for policymakers, urban planners, and law enforcement agencies in formulating strategies to improve road safety and reduce the incidence of accidents.

```python
#Importing the necessaries libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
data = pd.read_csv('C:/Users/MANOJ S/Downloads/archive (31)/Road Accident Data.csv')
data
```

#### Data Visualization

### Visualize the Distribution of Accidents by Day of the Week

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_Week', data=data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribution of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

### Visualize the Distribution of Accidents by Weather Conditions

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Conditions', data=data)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

 ### Visualize the Distribution of Accidents by Time of Day

```python
# Convert Time to datetime object
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M')

# Extract hour from Time
data['Hour'] = data['Time'].dt.hour

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Hour', bins=24, kde=True)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

###  Identify Accident Hotspots based on Latitude and Longitude

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', data=data, alpha=0.1)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### Visualize the Distribution of Accidents by Junction Control

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Junction_Control', data=data)
plt.title('Distribution of Accidents by Junction Control')
plt.xlabel('Junction Control')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

### Visualize the Relationship between Accident Severity and Number of Casualties

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Accident_Severity', y='Number_of_Casualties', data=data)
plt.title('Relationship between Accident Severity and Number of Casualties')
plt.xlabel('Accident Severity')
plt.ylabel('Number of Casualties')
plt.show()
```

### Visualize the Distribution of Accidents by Road Type

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Type', data=data)
plt.title('Distribution of Accidents by Road Type')
plt.xlabel('Road Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

### Visualize the Distribution of Accidents by Urban or Rural Area

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Urban_or_Rural_Area', data=data)
plt.title('Distribution of Accidents by Urban or Rural Area')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Number of Accidents')
plt.show()
```

### Visualize the Distribution of Accidents by Vehicle Type

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Vehicle_Type', data=data, order=data['Vehicle_Type'].value_counts().index)
plt.title('Distribution of Accidents by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.show()
```

### Conclusion:

#### Day of the Week:

The analysis of the distribution of accidents by day of the week reveals a higher occurrence of accidents on weekdays compared to weekends, which aligns with findings from previous studies. Weekdays, especially Friday, seem to have a higher number of accidents, possibly due to increased traffic congestion during commuting hours. However, it's essential to note that the type of accidents and other factors may influence this distribution.

#### Weather Conditions:

The visualization of accidents by weather conditions indicates that adverse weather conditions such as rain, snow, and fog contribute to a higher number of accidents. Combining bad weather with high winds exacerbates the situation, leading to more accidents. This underscores the importance of adjusting driving behavior and taking necessary precautions during inclement weather conditions to reduce the risk of accidents.

#### Time of Day:

Analyzing accidents by the time of day highlights a peak in accidents between 4 pm and 7 pm, coinciding with rush hour traffic. Another possible peak is observed around noon. Fewer accidents occur during late night hours, between 10 pm and 6 am. This pattern reflects the influence of factors like increased traffic volume and driver fatigue on accident occurrence throughout the day.

#### Junction Control:

The graph indicates that the most common type of junction control associated with accidents is "Give way or uncontrolled", followed by "Auto traffic signal". However, due to missing or out-of-range data for some junction control types, definitive conclusions are challenging. Further analysis considering total accident numbers and traffic volume at different types of junctions is necessary for a comprehensive understanding.

#### Accident Severity and Number of Casualties:

There appears to be a positive correlation between accident severity and the number of casualties, as evidenced by the graph. This suggests that as the severity of an accident increases, the number of casualties also tends to increase. However, correlation does not imply causation, and other factors may influence both accident severity and casualty numbers.

#### Road Type:

Single carriageways and one-way streets emerge as the most common locations for accidents based on the graph. However, without considering the total number of accidents, road length, and traffic volume, definitive conclusions about accident prevalence on different road types are limited.

#### Urban or Rural Area:

The graph indicates a higher number of accidents in urban areas compared to rural areas. While this suggests urban areas are more prone to accidents, factors such as population density and total distance traveled by drivers in each area need consideration for a comprehensive analysis.

#### Vehicle Type:

Motorcycles over 500cc are the most common type of vehicle involved in accidents, followed by cars, according to the graph. However, without accounting for total accidents and the number of vehicles of each type on the road, conclusions about the relative risk of different vehicle types in accidents are limited.

In conclusion, this exploratory data analysis provides valuable insights into the patterns and factors associated with traffic accidents. Understanding these trends can inform targeted interventions and policies aimed at enhancing road safety and reducing the incidence of accidents. However, further research and analysis are necessary to gain a comprehensive understanding of the complex dynamics underlying traffic accidents.
